```python
import pandas as pd
```

```python
penguins = pd.read_csv("https://github.com/mbrudd/csci290/raw/refs/heads/main/data/penguins.csv")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from scipy.spatial import distance
from sklearn.svm import SVC
```

```python
df = penguins.dropna()
```

```python
label_encoders = {}
```

```python
for column in ['species', 'island', 'sex']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
X = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'island', 'sex']]
y = df['species']
```

```python
class KNearestNeighbors:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predictions = []
        for x in X:
            distances = distance.cdist([x], self.X_train, 'euclidean')[0]
            k_indices = np.argsort(distances)[:self.k]
            k_nearest_labels = [self.y_train[i] for i in k_indices]
            most_common = np.bincount(k_nearest_labels).argmax()
            predictions.append(most_common)
        return np.array(predictions)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=42)
```

```python
knn = KNearestNeighbors(k=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'k-NN Accuracy: {accuracy:.2f}')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f'Logistic Regression Accuracy: {accuracy_log_reg:.2f}')
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f'SVM Accuracy: {accuracy_svm:.2f}')
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
```

```python
softmax_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=200)
softmax_reg.fit(X_train, y_train)
```

```python
y_pred_softmax = softmax_reg.predict(X_test)
accuracy_softmax = accuracy_score(y_test, y_pred_softmax)
print(f'Softmax Regression Accuracy: {accuracy_softmax:.2f}')
```

```python
sgd_clf = SGDClassifier(max_iter=1000, tol=1e-3, loss='log_loss')
```

```python
sgd_clf.fit(X_train, y_train)
```

```python
y_pred_sgd = sgd_clf.predict(X_test)
```

```python
accuracy_sgd = accuracy_score(y_test, y_pred_sgd)
```

```python
print(f'SGD Classifier Accuracy: {accuracy_sgd:.2f}')
```
